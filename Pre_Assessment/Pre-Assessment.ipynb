```python
### Q1. Loading data into Pandas
import pandas as pd
movie = pd.read_csv("movie.csv")
```

```python
### Q2. Writing an apply function that transforms a column
movie['Genre'] = movie['Genre'].apply(str.capitalize)
```

```python
### Q3. A basic Data Visualization using Seaborn or Plotly library or Matplotlib
import matplotlib.pyplot as plt
movie['Adjusted_Gross'] = movie['Adjusted_Gross'].replace(",","",regex=True).astype(float)
plt.scatter(x=movie['IMDb_Rating'], y=movie['Adjusted_Gross'])
plt.xlabel('IMDb Rating(0-10)')
plt.ylabel('Movie Gross Revenue')
plt.title('Movie Gross Revenue V.S. IMDb Rating')
plt.show()
```

```python
### Q4. Some type of Machine Learning technique on the data
import statsmodels.api as sm
x = movie["IMDb_Rating"]
y = movie["Adjusted_Gross"]
ols = sm.OLS(y, x).fit()
# predict movie gross revenue when it scores 9 on IMDb Rating
ols.predict(9)
```
